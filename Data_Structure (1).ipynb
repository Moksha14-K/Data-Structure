                         DATA STRUCTURE
1.What are data structures, and why are they important?
Data Structures are a specialized format for organizing, processing, retrieving and storing data. They are the fundamentals of any programming language around which a program is built. Types of data structures are like lists, tuples, dictionaries, etc.
They are important in Python because they help programmers solve complex problems and optimize algorithms.

2. Explain the difference between mutable and immutable data types with examples.
Data types in Python are categorized into mutable and immutable data types. Mutable data types are those whose values can be changed, whereas immutable data types are ones in which the values can’t be changed. We can add, remove, or modify the contents of a mutable data type but can’t do the same with an immutable data type.

3.What are the main differences between lists and tuples in Python?
The main difference between lists and tuples in Python is that lists are mutable, while tuples are immutable. Lists support a wider range of methods, such as appending, inserting, and removing elements. Tuples provide basic operations that allow you to retrieve information, like the length or index of a specific element. However , tuples are slightly faster than lists, especially when looking up values.

4.Describe how dictionaries store data.
A Python dictionary is a data structure that stores the value in key: value pairs. Values in a dictionary can be of any data type and can be duplicated, whereas keys can’t be repeated and must be immutable.

5.Why might you use a set instead of a list in Python?
Lists and Sets are standard Python data types that store values in a sequence. The major difference is that sets, unlike lists , cannot have multiple occurrences of the same element and store unordered values. Because sets cannot have multiple occurrences of the same element, it makes sets highly useful to efficiently remove duplicate values from a list and to perform common math operations like unions and intersections.

6.What is a string in Python, and how is it different from a list?
A String is a sequence of characters enclosed in a single or double quotes. On the other hand, list is a collection of items that can be of any data type and can be modified (mutable). Lists are represented by square brackets and can contain multiple elements separated by commas. Strings are immutable, meaning their values cannot be changed once they are created. While both lists and strings can be accessed using indexing and slicing, lists provide more flexibility as they can be modified, appended, or extended, while strings cannot be altered after creation.

7.How do tuples ensure data integrity in Python?
Tuples are immutable to ensure that their contents remain constant throughout their lifecycle, guaranteeing data integrity and reliability. This immutability allows tuples to be used as keys in dictionaries and elements in sets, as they can be hashed.

8.What is a hash table, and how does it relate to dictionaries in Python?
A hash table is a data structure that stores key-value pairs, where each key's hash value determines the correct bucket for that data. Hash tables are useful because they allow for quick access to data, regardless of the number of elements stored in the table.
Python dictionaries are implemented as hash tables. The keys in a dictionary are hashable, meaning they're generated by a hash function that produces a unique result for each unique value.

9.Can lists contain different data types in Python?
A List in Python can contain different types of data. It contains a mix of data types including strings, floats, and booleans.

10.Explain why strings are immutable in Python.
Strings are made immutable so that programmers cannot alter the contents of the object (even by mistake). This avoids unnecessary bugs.
       
11. What advantages do dictionaries offer over lists for certain tasks?
Dictionaries have two parts to every entry the key and the value and these are stored as pairs. This can give you more flexibility when storing data. The primary advantage would be retrieving stored values as it is much simpler to search through a dictionary.

12. Describe a scenario where using a tuple would be preferable over a list.
Tuples are an immutable data structure where the elements won't change after creation.They are generally faster and more memory-efficient than lists, making them better for fixed collections, especially as dictionary keys or when iteration speed is crucial. One such scenario where tuple would be preferable over a list is - Updating Aadhar card details of the employees as the details cannot be changed/immutable hence tuple would be better to use.

13. How do sets handle duplicate values in Python?
Set in python are immutable (unchangeable), They do not have duplicate values.They are unordered data structure i.e they can appear in a different order every time it is used.

14. How does the “in” keyword work differently for lists and dictionaries?
In Lists , "in" keyword works on the elements of the list whereas in dictionaries , "in" keyword searches the key pair and not value pair.

15. Can you modify the elements of a tuple? Explain why or why not.
No, you cannot modify the elements of a tuple in Python because tuples are "immutable," meaning once created, their elements cannot be changed, added, or removed.

16. What is a nested dictionary, and give an example of its use case?
A nested dictionary is a collection of dictionaries within another dictionary. This can be extended to multiple levels, creating a hierarchy of dictionaries. They represent data with multiple layers of organization. One such example of nested dictionary is Courses which has many sub courses like python , tableau , sql along with their duration and the fees.

17.Describe the time complexity of accessing elements in a dictionary.
The time complexity of accessing an element in a Python dictionary is O(1) on average: Python dictionaries use a hash table internally, which allows for constant time lookup. This means that the time it takes to retrieve an element from a dictionary is constant and does not depend on the size of the dictionary.

18. In what situations are lists preferred over dictionaries?
Lists are ordered, while dictionaries are unordered. Lists can store any data type, while dictionary keys can be of any immutable data type and values can be of any data type. Lists are faster for ordered operations like sorting, while dictionaries are faster for lookup operations.

19.Why are dictionaries considered unordered, and how does that affect data retrieval?
In python, dictionaries are un-ordered because Python dictionaries are not intended to be in order, as simple as that. If we want to collect a set of objects in order, we have only one choice of accessing them: through index.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
The difference between list and dictionary in Python lies in their structure; while a list is an ordered collection indexed by position and allowing duplicates, a dictionary is an unordered collection of unique key pairs.
Lists are best for sequential data, while dictionaries are tailored for data retrieval by specific identifiers. Data retrieval is fast in dictionaries whereas it is slow in lists.
